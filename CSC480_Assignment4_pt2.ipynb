```python
#target_book is set to None to act as an optional parameter.
#if target_book is supplied, generate single prediction for target user on target_book
def predict(train_d,target,K, target_book = None):
    pos_correlation = False
    #copy=true for dictionary
    df_correlation = pd.DataFrame(train_d,copy=True)
    correlation = []
    for i in range(len(df_correlation)):
        #for every row compute correlation between existing user and target user by finding books they both rated
        x = df_correlation.iloc[i]
        y_rating = []
        x_rating = []
        for k in range(len(target)):
            #add to list if existing and new user both rated the book
            if np.isnan(x[k]) == False and np.isnan(target[k]) == False:
                y_rating.append(target[k])
                x_rating.append(x[k])
        if len(x_rating) >= 2:
            #compute pearson correlation coefficient if number of overlapping ratings is at least 2
            correlation.append(np.corrcoef(x_rating,y_rating)[0,1])
        else:
            correlation.append(np.nan)

    df_correlation['p_correlation'] = correlation
    df_correlation.sort_values(by="p_correlation",ascending=False,inplace=True)
    

    #only considering positive correlation coefficients - sort descending and update K to slice out negative values
    if pos_correlation == False:
        if df_correlation.iloc[K - 1]['p_correlation'] > 0:
            pos_correlation = True
        else:
            K -= 1
    
    y_row = pd.Series(target,copy=True)
    books = df_correlation.columns.drop(labels='p_correlation')
    predictions = pd.Series(y_row, copy=True)

    for book in books:
        num_ = 0
        denom_ = 0
        #traverse the top K rows and calculate predicted rating
        for index, row in df_correlation.iloc[0:K].iterrows():
            if np.isnan(row.loc[book]) == False:
                #sum of correlations
                denom_ += row['p_correlation']
                #sum the product of rating and correlation WA
                num_ += row.loc[book] * row['p_correlation']
        if num_:
            avg_rate = num_/denom_
            predictions.update(pd.Series([avg_rate],index=[book]))
        else:
            #predict r according to avg if no overlap
            average = train_d[book].sum()
            avg = average/len(train_d[book])
            predictions.update(pd.Series([avg], index=[book]))
    #If target_book is supplied as a parameter, compute the single predicted rating of target_user on target_item
    if target_book:
        return predictions.loc[target_book]
    #Otherwise return predictions to to recommend/testing function to generate N recommendation/predictions
    else:    
        return predictions
```

```python
def recommend(train_d,target,K,N):
    target_recs = pd.Series(target,copy=True)
    pred_ratings = predict(train_d,target_recs,K)
    for index, value in target_recs.items():
        #get N predicted ratings for unrated movies only - drop out movies already rated
        if np.isnan(value) == False:
            pred_ratings[index] = np.nan
    pred_ratings.dropna(inplace=True)
    pred_ratings.sort_values(ascending=False)
    #Get at most N predictions based on Knn prediction, less if the user has fewer unrated items than N
    pred_ratings = pred_ratings[0:min(N,len(pred_ratings))]
    return pred_ratings
```

```python
#returns MAE, option to print comparison between test pred. and targets, option to print MAE
def testing_r(train_data,test_data,K,print_comparison=False, printMAE = False):
    MAE_num = 0
    MAE_denom = len(test_data)
    average_MAE = []
    MAE = []
    for index, row in test_data.iterrows():
        if print_comparison == True:
            print("\nUser: NU{}".format(index-20))
        pred_r = predict(train_data,row,K)
        for i, prediction in pred_r.items():
            if np.isnan(row.loc[i]) == False:
                if print_comparison == True:
                    print("Book: {},  Prediction: {},  Actual: {}".format(i,prediction,row.loc[i]))
                #calculate diff between prediction and actual rating - take absolute value
                r_diff = prediction - row.loc[i]
                #print(prediction)
                MAE_num += abs(r_diff)
               
        #compute MAE of instance and append to list of all MAE's of test data
        #computer total average MAE on test data
        #average_MAE.append(MAE)
    if printMAE:
        print("MAE = {}".format(np.sum(MAE)/len(test_data)))
    total_avg_MAE = np.mean(MAE)
    return total_avg_MAE
```

### Finding best K and then generating predictions for books unrated by NU1 and NU2 using best value of K

```python
MAE_vals = []
print("Finding K with minimum MAE:")
#Set parameter 'printMAE = True' to see progression of MAE values corresponding to values of K = 1-19
for k_val in range(1,len(training_users)-1):
    MAE_vals.append(testing_r(training_users,newUsers,k_val,print_comparison = False, printMAE=True))

books = data.columns
#finding the K with the minimum MAE value
min_k_mae = MAE_vals.index(min(MAE_vals)) + 1
print("K = {} with the minimum MAE".format(min_k_mae))
print("\nPredicted ratings for NU1 with best K:\n{}\n".format(recommend(training_users,newUsers.iloc[0], len(books),min_k_mae)))
print("Predicted ratings for NU2 with best K:\n{}\n".format(recommend(training_users,newUsers.iloc[1], len(books),min_k_mae)))
```
